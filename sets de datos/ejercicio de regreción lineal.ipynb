```python
#dummies o one-hot encoding eso lo que quiere decir es que si se tiene 
#NEAR BAY |INLAND |NEAR OCEAN| SE CREARA UNA COLUMNA PARA CADA UNO Y SI POR EJEMPLO los registros son nearbay
#    1         0        0
#    0         0        1

#es decir si es nerbay tendra el valor de 1 y el resto de 0 pero si es nerar ocean el tendra el valor de 1 y el resto de 0
#de esta manera es que se debe hacer para que no de un valor mayor y la ia no aprende  dandole un valor de 5 a algo que no tien erelevancia 
#esto se hace con pandas con la propiedad de get_dummies
```

```python
pd.get_dummies(datos_na["ocean_proximity"])
#aquí se optiene los dummies de los datos na en la column a ocean_proximitu y  devuelve los datos tal y ocmo aparecen si da un dato true o false, se le debe poner que me de un dato entero 
```

```python
dummies = pd.get_dummies(datos_na["ocean_proximity"],dtype=int)
```

```python
#se agrega la variable dommies a la tabla 
datos.join(dummies)
```

```python
#separar las caracteristicas de la etiqueta 
x= datos_na.drop(["median_house_value"], axis=1) #quitar esa columna
y= datos_na["median_house_value"] 
```

```python
#separar los datos en 2 partes: conjunto de entrenamiento y conjunto o set de pruebas

from sklearn.model_selection import train_test_split
#este me sirve para separar los datos, es decir es una función la cual separará los datos 
x_ent, x_pru,y_ent, y_pru =  train_test_split(x,y, test_size=.2)
#nombre de la función y sele dice cuanto debe separara para las pruebas esto regresa 4 variables  las cuales de eben guardaren variables
```

```python
#ent seran los datos de entrenamiento  pru seran los datos de prueba y el resultado de esto seran los registros y como están divividos 
x_ent.shape
```

```python
x_pru.shape
#aqui igual contiene 4087 datos de prueba divididos en 14 columnas, se debe tener en cuenta
#que estos datos de pruena son los que se han separado, es decir este es exeptuando los datos median_house_value
```

```python
y_pru.shape
#aquí solo mes muestra 4087 porque solo es la columna median_house_value
```

```python
from sklearn.linear_model import LinearRegression
#crear el modelo
modelo =LinearRegression()
```

```python
modelo.fit(x_ent,y_ent)
```

```python
predicciones = modelo.predict(x_pru)
```

```python
predicciones
#devuelve un array
```

```python
pd.DataFrame(predicciones)
```

```python
#agregar a un diccionario
comparativa = {"predicciones" :predicciones}
pd.DataFrame(comparativa)
```

```python
#agregar el valor real es decir los datos de prueba
comparativa = {"Predicciones": predicciones, "Valor Real": y_pru}
pd.DataFrame(comparativa)
```

```python
#sobre ajuste o Overfitting
print(modelo.score(x_ent, y_ent))
print(modelo.score(x_pru, y_pru))
```

```python
#error
from sklearn.metrics import mean_squared_error
import numpy as np

mse = mean_squared_error(y_pru, predicciones) 
#mean_squared_error toma todos los datos,sca la diferencia y la eleva al cuadrado 
```

```python
mse
```

```python
#como es un número muy grande, se saca raiz, esta raiz es para poder tener el número ya que se esta elevando al cuadrado ya que existen números negativos
rmse = np.sqrt(mse)
```
