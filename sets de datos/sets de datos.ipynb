con la funci√≥n dropna

```python
datos_na = datos.dropna()
```

```python
datos_na.info()
```

ahora se va a conveertir la caracteristica categ√≥rica a n√∫merica
es decir la variable de proximidad al oceano


```python
datos_na["ocean_proximity"]
```

```python
datos_na["ocean_proximity"].value_counts()
#se puede dar el valor de 1,2,3,4,5 pero esto puede ser malo ya que el modelo puede darle m√°s valor al n√∫mero 5 que 1 esto no queremos que lo aprenda el modelo por lo cual hay algo llamado dummies o one-hot encoding
```

```python
#dummies o one-hot encoding eso lo que quiere decir es que si se tiene 
#NEAR BAY |INLAND |NEAR OCEAN| SE CREARA UNA COLUMNA PARA CADA UNO Y SI POR EJEMPLO los registros son nearbay
#    1         0        0
#    0         0        1

#es decir si es nerbay tendra el valor de 1 y el resto de 0 pero si es nerar ocean el tendra el valor de 1 y el resto de 0
#de esta manera es que se debe hacer para que no de un valor mayor y la ia no aprende  dandole un valor de 5 a algo que no tien erelevancia 
#esto se hace con pandas con la propiedad de get_dummies
```

```python
pd.get_dummies(datos_na["ocean_proximity"])
#aqu√≠ se optiene los dummies de los datos na en la column a ocean_proximitu y  devuelve los datos tal y ocmo aparecen si da un dato true o false, se le debe poner que me de un dato entero 
```

```python
dummies = pd.get_dummies(datos_na["ocean_proximity"],dtype=int)
```

```python
#se agrega la variable dommies a la tabla 
datos.join(dummies)
```

```python
#se asigna a la variable
datos_na = datos_na.join(dummies)
```

```python
datos_na.head()
```

```python
#eliminar ocean_proxomity con esto se elimina la columna que ya no es necesaria 
datos_na= datos_na.drop(["ocean_proximity"], axis=1)
```
