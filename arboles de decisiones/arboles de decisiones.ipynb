```python
#entrenamiento
x=datos.drop(["Survived"],axis=1)
y=datos["Survived"]
```

```python
datos.head()
```

```python
from sklearn.model_selection import train_test_split

x_ent, x_pru , y_ent, y_pru = train_test_split(x, y, test_size=.2)
```

```python
#comienza el arbol de desicion
from sklearn.tree import DecisionTreeClassifier
modelo = DecisionTreeClassifier(max_depth=50)
#este me sirve para clasificar  quien sibrevivio o no
#con criterion= se le dice los criterios que quiere que utilice esto va dentro de los parametros, es decir dentro de (criterion="") por defecto es gini
#también se le puede decir max_depth que esto es que tran profundo puede ser el árbol hacia abajo entre más profundo puede tener más opciones de desicion pero si es demasiado profundo puede llegar al sobre ajuste se ysa de la siguiente manere (max_depth=50)
modelo.fit(x_ent,y_ent)
#se entrena al modelo
predicciones= modelo.predict(x_pru)
```

```python
#como le fue al modelo
from sklearn.metrics import accuracy_score

accuracy_score(y_pru, predicciones)
#se hace un quiz o se evalua como le fue al modelo enviandole las predicciones y las respuestas
```

```python
from sklearn.tree import DecisionTreeClassifier

resultados =[]
for i in range(1,15):
    modelo = DecisionTreeClassifier(max_depth=50)
    modelo.fit(x_ent,y_ent)
    predicciones= modelo.predict(x_pru)
    exactitud = accuracy_score(y_pru, predicciones)
    print(f"resultado para {i} es {exactitud}")
    resultados.append(exactitud)
```

```python
sb.lineplot(data=resultados)
```

```python
modelo = DecisionTreeClassifier(max_depth=8)
modelo.fit(x_ent,y_ent)
predicciones= modelo.predict(x_pru)
exactitud = accuracy_score(y_pru, predicciones)
```
